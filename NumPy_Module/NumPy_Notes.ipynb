# Declaring a Numpy Array

#### Create an array, each element of which is zero with data type = float64

```python
z = np.zeros((9,2)) # Makes an array of 9 x 2
print(type(z[0][0]))
z
```

#### Create an array, each element of which is 1 with data type = float64

```python
z = np.ones((6,5), np.int32)
print(type(z[0][0]))
z
```

#### Number of dimensions in the array

```python
print(z.ndim)
z.shape
```

#### shape : Returns =  tuple of ints,  (height, width, depth)

```python
z.shape = (10, 3)
z
```

#### np.arange(a, b, c) makes array of values from 10 to 85 with gap of c

#### Dot Product Operations - Matrix Mutiplication

```python
a = np.random.permutation(9).reshape((3, 3))
print(f"a = {a}")
b = np.random.permutation(9).reshape((3, 3))
print(f"b = {b}")
print("------------------------------")
print(a @ b)
print(np.dot(a, b))
```

#### Element Wise Matrix Multiplication 

```python
a = np.random.permutation(9).reshape((3, 3))
print(f"a = {a}")
b = np.random.permutation(9).reshape((3, 3))
print(f"b = {b}")
print("------------------------------")
print(a * b)
```
